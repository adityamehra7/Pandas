Making changes to the data

```python
# adding a new column by summing up some of the existing ones

df['Total'] = df['HP'] + df['Attack'] + df['Defense'] + df['Speed']
```

```python
df.head(5)
```

```python
df['Total'] = df['HP'] + df['Total'] + df['Sp. Def'] + df['Sp. Atk']
```

```python
df.head(5)
```

```python
df.sort_values('Total',ascending=False)
```

```python
# dropping a column/columns from a dataframe

df = df.drop(columns=['Total'])
```

```python
# adding in another way

df['Total'] = df.iloc[:,4:10].sum(axis=1)
```

```python
df.head(5)
```

```python
# CHANGING THE POSITION OF A SPECIFIC COLUMN

cols = list(df.columns.values)
df = df[ cols[0:4] + [cols[-1]] + cols[4:12] ]
df.head(5)
```
